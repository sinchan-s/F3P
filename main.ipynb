```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_theme()
```

```python
df = pd.read_csv("Physical data with coverage.csv")
df
```

```python
#spec_article = df[df[['Article No.']=='A150A567',df['Weave']=='2/1 S-TWILL']]
spec_article = df.loc[(df['Article No.']=='A150A567')&(df['Finish']=='CT+SF')]
spec_article
```

```python
epi_ppi = spec_article[['EPI','PPI']].sort_values(by='EPI')
epi_ppi
```

```python
x, y = epi_ppi.EPI.values, epi_ppi.PPI.values
x
```

```python
fig = plt.figure()
plt.plot(x, y, "bo")
plt.xlabel("EPI", fontsize=20)
plt.ylabel("PPI", fontsize=20)
```
